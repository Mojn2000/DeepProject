## Performance curves


```python
def tpr(mat):
  return (mat[1,:]/(mat[1,:]+mat[2,:]))  # TP / (TP+FP)

def tnr(mat):
  return (mat[4,:]/(mat[3,:]+mat[4,:]))  # TN / (TN+FN)

plt.rcParams.update(plt.rcParamsDefault)
plt.figure(figsize=(8, 6), dpi=800)
plt.plot([0,1],[1,0],linestyle = 'dashed')
plt.plot(tpr(v1_3),tnr(v1_3),marker='.')
plt.plot(tpr(v1_3_r),tnr(v1_3_r),marker='.')
plt.plot(tpr(v1_1),tnr(v1_1),marker='.')
plt.plot(tpr(v1_1_r),tnr(v1_1_r),marker='.')
plt.plot(tpr(v2_0),tnr(v2_0),marker='.')
plt.xlabel('True positive rate', fontsize = 16)
plt.ylabel('True negative rate',fontsize = 16)
plt.title('Performance curve', fontsize = 20)
plt.grid()
plt.legend(['Random guess','Uni-directional','Robust uni','Bi-directional', 'Robust bi','Supervised'], loc = 'lower left')
plt.show()

```
